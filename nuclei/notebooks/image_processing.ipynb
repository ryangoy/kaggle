```python
import numpy as np
from os import listdir, remove, rmdir, rename, makedirs
from os.path import isdir, join, exists
from scipy.misc import imread
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
dataset_path = '/home/ryan/cs/datasets/nuclei'
train_path = join(dataset_path, 'stage1_train')
test_path = join(dataset_path, 'stage1_test')

resized_train_path = join(dataset_path, 'resized_train')
resized_test_path = join(dataset_path, 'resized_test')

WIDTH = 256
HEIGHT = 256

if not isdir(resized_train_path):
    makedirs(resized_train_path)
if not isdir(resized_test_path):
    makedirs(resized_test_path)
```

```python
def create_combined_masks(path):
    samples = sorted(listdir(path))
    
    for sample in samples:
        sample_path = join(path, sample)
        if isdir(sample_path):
            print sample
            image_path = join(sample_path, 'images', sample+'.png')
            if not exists(image_path):
                print('Error,  image does not exist.')
            else:
                masks_path = join(sample_path, 'masks')
                total_mask = None
                for mask_name in sorted(listdir(masks_path)):
                    mask_path = join(masks_path, mask_name)
                    mask = imread(mask_path)
                    if total_mask is not None:
                        total_mask += mask
                    else:
                        total_mask = mask
                plt.imshow(total_mask)
                plt.show()
                break
                
def resize_data(path):
    samples = sorted(listdir(path))
    max_diff = 0
    dims  = None
    counter = 0
    for sample in samples:
        sample_path = join(path, sample)
        if isdir(sample_path):
            image_path = join(sample_path, 'images', sample+'.png')
            if not exists(image_path):
                print('Error,  image does not exist.')
            else:
                
                img = imread(image_path, mode='RGB')
                
                # Convert to grayscale with a naive method
                grayscale = np.mean(img, axis=2)
                
                # Invert images with bright backgrounds
                if np.mean(grayscale) > 128:
                    grayscale = 255*np.ones(grayscale.shape) - grayscale
                    
                num_pixels = img.shape[0] * img.shape[1]
                thresh_bot = int(0.05 * num_pixels)
                thresh_top = int(0.99 * num_pixels)
                
                lb = np.partition(grayscale.ravel(), thresh_bot)[thresh_bot]
                ub = np.partition(grayscale.ravel(), thresh_top)[thresh_top]
                
                grayscale[grayscale < lb] = lb
                grayscale[grayscale > ub] = ub
                    
                # Color balance
                norm_img = cv2.normalize(grayscale, dst=None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
                
                
                x_ds = img.shape[0] / WIDTH
                y_ds = img.shape[1] / HEIGHT

                # Assumes x and y dimension are relatively similar
                if x_ds > 1.0 or y_ds < 1.0:
                    # Gaussian filter and downsample
                    x_std = x_ds - 1
                    y_std = y_ds - 1
                    x_ksize = int((x_ds+1)/2)*2+1
                    y_ksize = int((y_ds+1)/2)*2+1
                    dst = cv2.GaussianBlur(norm_img, (x_ksize, y_ksize), x_std, y_std)
                    resized_img = cv2.resize(dst, (WIDTH, HEIGHT), interpolation=cv2.INTER_LANCZOS4)
                else:
                    # Upsample
                    resized_img = cv2.resize(norm_img, (WIDTH, HEIGHT), interpolation=cv2.INTER_LANCZOS4)
                
                
                    
                    
                # Ideally, we want the color distribution to be two-humped, one hump near 0 intensity
                # corresponding to the background, and one hump near 1 intensity corresponding to nuclei.
                # Saturation 
                
                plt.imshow(img)
                plt.show()
                
                # 70 is a decent threshold for nuclei / background
#                 resized_img[resized_img < 70] = 0
#                 resized_img[resized_img >= 70] = 255
                plt.imshow(resized_img)
                plt.show()
                plt.hist(resized_img.ravel(),256,[0,256])
                plt.ylim(0,5000)
                plt.show()
                counter += 1
                if counter > 10:
                    break
                    
                
                
                
```
