```python
import pickle
import os
import numpy as np
import pandas as pd
```

```python
fish_path = '/home/ryan/cs/datasets/ncfm/test/'
predictions = 'predictions.pkl'
save_path = 'submission_ssd_1.csv'
```

```python
predictions = pickle.load(open(predictions, 'r'))
num_images = len(predictions)
num_categories = 8
```

```python
formatted = np.zeros((num_images, num_categories))
```

```python
classes = ['ALB', 'BET', 'DOL', 'LAG', 'NoF', 'OTHER', 'SHARK', 'YFT']
# we need a mapping from the 6 class classifier to the submission classes
index_map = {1:0, 2:1, 3:2, 4:3, 5:6, 6:7}
res_list = []
for i in range(num_images):
    pred = predictions[i]
    conf_values = pred[:,1][:3]
    labs = pred[:,0][:3]
    res = [(1-conf_values[0])/10] * 8
    for j in range(len(labs)):
        res[index_map[labs[j]]] = conf_values[j]
    if conf_values[0] < 0.6:
        res[4] = 6
        res[5] = 3
    res_list.append(res)
```

```python
sorted_names = sorted(os.listdir(fish_path))
print sorted_names[:10]
```

```python
res_list = np.array(res_list)
```

```python
print res_list.shape
print np.array(sorted_names).shape
print len(classes)
```

```python
df = pd.DataFrame(data = res_list, index = sorted_names, columns = classes)
```
