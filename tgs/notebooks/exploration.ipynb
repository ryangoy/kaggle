```python
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.misc import imread
from scipy import ndimage
# %matplotlib inlin
```

```python
TRAIN_IMAGES_PATH = '/home/ryan/cs/datasets/tgs/train/images'
TEST_IMAGES_PATH = '/home/ryan/cs/datasets/tgs/test/images'
TRAIN_MASKS_PATH = '/home/ryan/cs/datasets/tgs/train/masks'
```

```python
train_images = os.listdir(TRAIN_IMAGES_PATH)
train_images = sorted(train_images)
```

```python
subset_images = train_images[5:10]
```

```python
def plot(data, title):
    plot.i += 1
    plt.subplot(2, 2,plot.i)
    plt.imshow(data)
    plt.title(title)
```

```python
plot.i = 0
for img_name in subset_images:
    img = imread(os.path.join(TRAIN_IMAGES_PATH, img_name))
    mask = imread(os.path.join(TRAIN_MASKS_PATH, img_name))
    gauss_lowpass = ndimage.gaussian_filter(img, 5)
    gauss_highpass = img - gauss_lowpass
    plt.figure()
    plot.i = 0
    plot(img, img_name)
    plot(gauss_highpass/10, 'Highpass')

    plot(gauss_lowpass, 'Lowpass')
    plot(mask, 'Mask')
plt.show()
```
