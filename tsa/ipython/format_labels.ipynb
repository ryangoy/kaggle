```python
import numpy as np
import pandas as pd
from os import listdir
from os.path import join
```

```python
raw_csv = pd.read_csv('/home/ryan/cs/datasets/tsa/stage1/labels.csv')
```

```python
raw_csv[20:50]
```

```python
num_zones = 17
new_df = pd.DataFrame(columns=('Id', 'Label'))
curr_name = ''
curr_count = 0
curr_image = 0

for index, row in raw_csv.iterrows():
    # same image, different zone
    if row['Id'].split('_')[0] == curr_name and curr_count < num_zones:
        one_hot[int(row['Id'].split('_')[1][4:])-1] = row['Probability']
        curr_count += 1
    # start a new image
    elif row['Id'].split('_')[0] != curr_name and (curr_count == num_zones or curr_count == 0):
        # dump previous image
        if curr_count != 0:
            new_df.loc[curr_image] = [curr_name, one_hot]
            curr_image += 1
        
        one_hot = np.zeros(num_zones)
        one_hot[int(row['Id'].split('_')[1][4:])-1] = row['Probability']     
        curr_name = row['Id'].split('_')[0]      
        curr_count = 1
    else:
        print('Error: zones not consistent with path name.')
        break
    
        
```
