```python
images_path = '/home/ryan/cs/datasets/ncfm/train_all'
classes = {
               'ALB':  [1,0,0,0,0,0,0,0], 
               'BET':  [0,1,0,0,0,0,0,0], 
               'DOL':  [0,0,1,0,0,0,0,0],
               'LAG':  [0,0,0,1,0,0,0,0],
               'NoF':  [0,0,0,0,1,0,0,0], 
               'OTHER':[0,0,0,0,0,1,0,0],
               'SHARK':[0,0,0,0,0,0,1,0],
               'YFT':  [0,0,0,0,0,0,0,1],
          }
```

```python
def sloth_to_npy(raw_pickle):
    sloth_dict = {}
    for img_dict in raw_pickle:
        img = Image.open(os.path.join(images_path, os.path.basename(img_dict['filename'])))
        width, height = img.size

        for annotation_dict in img_dict['annotations']:
            xmin = annotation_dict['x']
            xmax = xmin + annotation_dict['width']
            ymin = annotation_dict['y']
            ymax = ymin + annotation_dict['height']
            data_array = [xmin/width, ymin/height, xmax/width, ymax/height] \
                         + classes[annotation_dict['class']]
            img_name = os.path.basename(img_dict['filename'])
            if img_name in sloth_dict:
                sloth_dict[img_name].append(data_array)               
            else:
                sloth_dict[img_name] = [data_array]
    return sloth_dict
```

```python
new_dict = sloth_to_npy(raw_pickle)
```
